# Wofür ist `__name__` gut? 
s. [[Source](https://www.freecodecamp.org/news/whats-in-a-python-s-name-506262fe61e8/)]

```python
print('The value of __name__ is ' + __name__)
```

Aber wenn wir `nntw.NeuralNetwork(...)` aufrufen, wird der Modul-Dateiname `'NeuralNetwork'` an `__name__` übergeben, sodass `if __name__=='__main__'` zu `False` auswertet:

> Within a module, the module’s name (as a string) is available as the value of the global variable `__name__`. [[Source](https://docs.python.org/3/tutorial/modules.html)]

```python
print('The value of __name__ is ' + nntw.__name__)
```

```python
nn = nntw.NeuralNetwork([1, 10, 10, 1],activations=['sigmoid', 'sigmoid', 'linear'])
X = 2*np.pi*np.random.rand(1000).reshape(1, -1)
y = np.sin(X)    

nn.train(X, y, epochs=2000, batch_size=16, lr = .5)
_, a_s = nn.feedforward(X)
#print(y, X)
plt.scatter(X.flatten(), y.flatten())
plt.scatter(X.flatten(), a_s[-1].flatten())
plt.show()
```
